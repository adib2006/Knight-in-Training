```python
df = df.sort_values(by='landed_name', ascending=False)
plt.figure(figsize=(90,10))
sns.countplot(x='landed_name', data = df)
```

```python
c = (((df.groupby(['landed_name']).mean())['time_landed']).sort_values( ascending=False))
b = ((df["landed_name"].value_counts()).index)

plt.figure(figsize=(90,10))
plt.bar(b,c)
plt.show
```

```python
c = (((df.groupby(['landed_name']).mean())['landed_location']).sort_values( ascending=False))
b = ((df["landed_name"].value_counts()).index)

plt.figure(figsize=(90,10))
plt.bar(b,c)
plt.show
```
